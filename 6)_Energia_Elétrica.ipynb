```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Gerando dados fictícios
np.random.seed(0)
num_aparelhos = np.random.randint(1, 10, size=100)
custo_energia = num_aparelhos * 50 + np.random.normal(0, 10, size=100)

# Criando um DataFrame
dados = pd.DataFrame({'num_aparelhos': num_aparelhos, 'custo_energia': custo_energia})

# Dividindo os dados em conjuntos de treino e teste
X = dados[['num_aparelhos']]
y = dados['custo_energia']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Criando e treinando o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Fazendo previsões
y_pred_train = modelo.predict(X_train)
y_pred_test = modelo.predict(X_test)

# Calculando o erro quadrático médio (MSE)
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)  # Corrigido aqui

print(f'MSE (Treino): {mse_train}')
print(f'MSE (Teste): {mse_test}')

# Visualizando os resultados
plt.scatter(X, y, color='blue', label='Dados Reais')
plt.plot(X, modelo.predict(X), color='red', label='Modelo de Regressão Linear')
plt.xlabel('Número de Aparelhos')
plt.ylabel('Custo de Energia (R$)')
plt.legend()
plt.show()
```
